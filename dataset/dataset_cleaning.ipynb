```python
import pandas as pd
import numpy as np
```

```python
drivers_df = pd.read_csv('archive/drivers.csv')
cols_with_backslashN = (drivers_df == "\\N").any()
print(f'Colonne con valori nulli:\n{cols_with_backslashN[cols_with_backslashN]}')
drivers_df.head()
```

```python
drivers_df.info()
```

```python
drivers_df['nationality'] = drivers_df['nationality'].replace({'Argentine': 'Argentinian'})
drivers_df.groupby('nationality').count()
```

```python
# Il code può essere ottenuto dalle prime tre lettere del cognome
cols_to_keep = [col for col in drivers_df.columns if '\\N' not in drivers_df[col].values]
drivers_df = drivers_df[cols_to_keep]
```

```python
drivers_df["dob"] = pd.to_datetime(drivers_df["dob"], errors="coerce")
drivers_df.head()
```

```python
drivers_df.info()
```

```python
drivers_df.to_csv('./cleaned/drivers_cleaned.csv', index=False)
```
